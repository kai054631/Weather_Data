```python
df=pd.read_csv("https://raw.githubusercontent.com/kai054631/Weather_Data/refs/heads/main/seattle_weather.csv")
print(df.shape) #Print the shape of the dataset
print(df.describe())
df.head()# Print First 5 Rows
```

```python
g=pd.get_dummies(df,columns=['weather'],dtype= int)
X = g.drop(columns=["precipitation"])
y = g["precipitation"]

X_train, X_test, y_train, y_test = split(X, y, test_size=0.25, random_state=42)
print(g.shape) #Print the shape of the dataset
print(g.describe())
g.head()
```

```python
#use spotchecking
models ={}
models['LR'] = LogisticRegression()
models['CART'] = DecisionTreeClassifier()
models['RF'] = RandomForestClassifier()
models['SVM'] = SVC()
models['KNN'] = KNeighborsClassifier()
models['GBM'] = GradientBoostingClassifier()

kf = KFold(n_splits=3, shuffle=True, random_state=42)

X_train,X_test,Y_train,Y_test=split(X,y,test_size=0.25,random_state=25)
kf = KFold(n_splits=3,shuffle=True,random_state=42)
for n in models:
  score=cross_val_score(models[n], X, y, cv=kf, n_jobs=-1)
  print(f'{n}:{score.mean():.3%},  {score.std():.3%}')
```
